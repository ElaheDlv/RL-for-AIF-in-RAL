```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Folder path ---
folder_path = './rgb_best_1_dev'  # change if needed

# --- Function to calculate average distance ---
def average_distance(df):
    if {'vehicle_location_x', 'vehicle_location_y', 'waypoint_location_x', 'waypoint_location_y'}.issubset(df.columns):
        distances = np.sqrt((df['vehicle_location_x'] - df['waypoint_location_x'])**2 + (df['vehicle_location_y'] - df['waypoint_location_y'])**2)
        return distances.mean()
    else:
        print("⚠️ Missing columns in file.")
        return None

# --- Loop through all CSV files that contain 'step_metrics' ---
for file_name in os.listdir(folder_path):
    if 'step_metrics' in file_name and file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        print(f"\n📄 Processing file: {file_name}")
        
        try:
            # Load data
            data = pd.read_csv(file_path)
            
            # --- Plot 1: (X, Y) Path ---
            if {'vehicle_location_x', 'vehicle_location_y'}.issubset(data.columns):
                plt.figure(figsize=(8, 6))
                plt.plot(data['vehicle_location_x'], data['vehicle_location_y'], 'o-', label='(X, Y) Path')
                plt.title(f'Agent Path: {file_name}')
                plt.xlabel('X Coordinate')
                plt.ylabel('Y Coordinate')
                plt.legend()
                plt.grid(True)
                plt.show()
            else:
                print("⚠️ Skipping (X, Y) plot: columns missing.")
            
            # --- Plot 2: (waypoint_x, waypoint_y) Path ---
            if {'waypoint_location_x', 'waypoint_location_y'}.issubset(data.columns):
                plt.figure(figsize=(8, 6))
                plt.plot(data['waypoint_location_x'], data['waypoint_location_y'], 'o-', label='(waypoint_x, waypoint_y) Path', color='orange')
                plt.title(f'Waypoint Path: {file_name}')
                plt.xlabel('X Coordinate')
                plt.ylabel('Y Coordinate')
                plt.legend()
                plt.grid(True)
                plt.show()
            else:
                print("⚠️ Skipping waypoint plot: columns missing.")
            
            # --- Compute average distance ---
            avg_dist = average_distance(data)
            if avg_dist is not None:
                print(f"✅ Average distance for {file_name}: {avg_dist:.3f}")
            else:
                print("⚠️ Average distance not computed (missing columns).")
        
        except Exception as e:
            print(f"❌ Error reading {file_name}: {e}")
```
